```python
# Import packages
from pyspark.sql import SparkSession
import time

# Create a SparkSession
spark = SparkSession.builder.appName("SparkSQL").getOrCreate()
```

```python
# 1. Read in the AWS S3 bucket into a DataFrame.
from pyspark import SparkFiles
url = "https://2u-data-curriculum-team.s3.amazonaws.com/dataviz-classroom/v1.2/22-big-data/home_sales_revised.csv"

spark.sparkContext.addFile(url)
df = spark.read.csv(SparkFiles.get("home_sales_revised.csv"), sep=",", header=True)
df.show()
```

```python
# 2. Create a temporary view of the DataFrame.

df.createOrReplaceTempView('home_sales')
```

```python
# 3. What is the average price for a four bedroom house sold per year, rounded to two decimal places?

avg_price_four_bed = spark.sql("SELECT YEAR(date) AS Year, ROUND(AVG(price),2) AS Average_Price FROM home_sales WHERE bedrooms==4 group by YEAR(date) ORDER BY YEAR(date) asc" )
avg_price_four_bed.show()
```

```python
# 4. What is the average price of a home for each year the home was built,
# that have 3 bedrooms and 3 bathrooms, rounded to two decimal places?
avg_price_built_year = spark.sql("SELECT date_built AS Year_Built, ROUND(AVG(price),2) AS Average_Price FROM home_sales WHERE bedrooms==3 and bathrooms==3 group by date_built order by date_built asc")
avg_price_built_year.show()
```

```python
# 5. What is the average price of a home for each year the home was built,
# that have 3 bedrooms, 3 bathrooms, with two floors,
# and are greater than or equal to 2,000 square feet, rounded to two decimal places?

avg_price_built_year1 = spark.sql("SELECT date_built AS Year_Built, ROUND(AVG(price),2) AS Average_Price FROM home_sales WHERE bedrooms==3 and bathrooms==3 and floors==2 and sqft_living>=2000 group by date_built order by date_built asc")
avg_price_built_year1.show()
```

```python
# 6. What is the average price of a home per "view" rating, rounded to two decimal places,
# having an average home price greater than or equal to $350,000? Order by descending view rating.
# Although this is a small dataset, determine the run time for this query.

start_time = time.time()

avg_price_view = spark.sql("SELECT view AS Rating, ROUND(AVG(price),2) AS Average_Price FROM home_sales WHERE price>=350000 group by view order by view desc")
avg_price_view.show()

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# 7. Cache the the temporary table home_sales.
spark.sql("cache table home_sales")
```

```python
# 8. Check if the table is cached.
spark.catalog.isCached('home_sales')
```

```python
# 9. Using the cached data, run the last query above, that calculates
# the average price of a home per "view" rating, rounded to two decimal places,
# having an average home price greater than or equal to $350,000.
# Determine the runtime and compare it to the uncached runtime.

start_time = time.time()

avg_price_view1 = spark.sql("SELECT view AS Rating, ROUND(AVG(price),2) AS Average_Price FROM home_sales WHERE price>=350000 group by view order by view desc")
avg_price_view1.show()

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# 10. Partition by the "date_built" field on the formatted parquet home sales data
df.write.parquet('home_sales', mode='overwrite')
df.write.partitionBy("date_built").mode("overwrite").parquet("home_sales")
```

```python
# 11. Read the formatted parquet data.
p_df=spark.read.parquet('home_sales')
```

```python
# 12. Create a temporary table for the parquet data.
p_df.createOrReplaceTempView('temp_home_sales')
```

```python
# 13. Using the parquet DataFrame, run the last query above, that calculates
# the average price of a home per "view" rating, rounded to two decimal places,
# having an average home price greater than or equal to $350,000.
# Determine the runtime and compare it to the cached runtime.

start_time = time.time()

avg_price_view2 = spark.sql("SELECT view AS Rating, ROUND(AVG(price),2) AS Average_Price FROM temp_home_sales WHERE price>=350000 group by view order by view desc")
avg_price_view2.show()

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# 14. Uncache the home_sales temporary table.
spark.sql("uncache table home_sales")
```

```python
# 15. Check if the home_sales is no longer cached

spark.catalog.isCached("home_sales")
```
